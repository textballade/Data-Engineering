```python
con.sql("""
SELECT table_schema, table_name
FROM information_schema.tables
""").df()
```

```python
con.sql("""
    SELECT * 
    FROM prod.fct_monthly_zone_revenue
    LIMIT 3    
""").show()
```

```python
con.sql("""
    SELECT SUM(total_monthly_trips) AS t_sum
    FROM prod.fct_monthly_zone_revenue
    WHERE service_type = 'Green' 
    AND (revenue_month >= '2019-10-01' AND revenue_month < '2019-11-01')
    ORDER BY t_sum DESC
    LIMIT 10
""").show()
```
